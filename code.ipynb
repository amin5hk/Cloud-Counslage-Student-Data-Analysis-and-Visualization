Que 3 : What is the distribution of students across different graduation year ? 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Data analyst Data.csv')

# Count the number of students for each graduation year
year_counts = df['Year of Graduation'].value_counts()

# Sort the counts by graduation year
year_counts = year_counts.sort_index()

# Plotting
plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Students across Different Graduation Years')
plt.xlabel('Graduation Year')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each bar with its count
for i, count in enumerate(year_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

Que 4 : What is the distribution of student's experience with python programming ? 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Data analyst Data.csv')

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(df['Experience with python (Months)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Students\' Experience with Python Programming')
plt.xlabel('Experience with Python (Months)')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each bin with its count
bin_counts = df['Experience with python (Months)'].value_counts(bins=20)
for i, count in enumerate(bin_counts):
    plt.text(bin_counts.index[i].mid, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

Que 6 : How does the average GPA among different colleges(Top 5) ?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Data analyst Data.csv')

# Calculate the average GPA for each college
college_gpa = df.groupby('College Name')['CGPA'].mean().sort_values(ascending=False).head(5)

# Plotting
plt.figure(figsize=(16, 10))
college_gpa.plot(kind='bar', color='skyblue')
plt.title('Average GPA among Different Colleges (Top 5)')
plt.xlabel('College Name')
plt.ylabel('Average GPA')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each bar with its GPA
for i, gpa in enumerate(college_gpa):
    plt.text(i, gpa, f'{gpa:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

Que 9 : Can we identify any relation between family income and GPA ? 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Data analyst Data.csv')

# Calculate the average GPA for each family income category
avg_gpa_by_income = df.groupby('Family Income')['CGPA'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['Family Income'], df['CGPA'], color='skyblue', alpha=0.5, label='Individuals')
plt.plot(avg_gpa_by_income.index, avg_gpa_by_income.values, color='red', marker='o', linestyle='-', label='Average GPA by Family Income')
plt.title('Relationship between Family Income and GPA')
plt.xlabel('Family Income')
plt.ylabel('GPA')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Que 11 : How does the expected salary vary based on factors like 'GPA', 'Family Income', 'Experience with python(Months)' ?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Data analyst Data.csv')

# Calculate the average expected salary for each factor
avg_salary_by_gpa = df.groupby('CGPA')['Expected salary (Lac)'].mean()
avg_salary_by_income = df.groupby('Family Income')['Expected salary (Lac)'].mean()
avg_salary_by_experience = df.groupby('Experience with python (Months)')['Expected salary (Lac)'].mean()

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Scatter plot for GPA vs Expected Salary
axs[0].scatter(df['CGPA'], df['Expected salary (Lac)'], color='skyblue', alpha=0.5)
axs[0].plot(avg_salary_by_gpa.index, avg_salary_by_gpa.values, color='red', linestyle='-', label='Average Salary')
axs[0].set_title('GPA vs Expected Salary')
axs[0].set_xlabel('GPA')
axs[0].set_ylabel('Expected Salary (Lac)')
axs[0].legend()
axs[0].grid(True, linestyle='--', alpha=0.7)

# Scatter plot for Family Income vs Expected Salary
axs[1].scatter(df['Family Income'], df['Expected salary (Lac)'], color='salmon', alpha=0.5)
axs[1].plot(avg_salary_by_income.index, avg_salary_by_income.values, color='red', linestyle='-', label='Average Salary')
axs[1].set_title('Family Income vs Expected Salary')
axs[1].set_xlabel('Family Income')
axs[1].set_ylabel('Expected Salary (Lac)')
axs[1].legend()
axs[1].grid(True, linestyle='--', alpha=0.7)

# Scatter plot for Experience with Python vs Expected Salary
axs[2].scatter(df['Experience with python (Months)'], df['Expected salary (Lac)'], color='green', alpha=0.5)
axs[2].plot(avg_salary_by_experience.index, avg_salary_by_experience.values, color='red', linestyle='-', label='Average Salary')
axs[2].set_title('Experience with Python vs Expected Salary')
axs[2].set_xlabel('Experience with Python (Months)')
axs[2].set_ylabel('Expected Salary (Lac)')
axs[2].legend()
axs[2].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

--Visualization shows us that higher the GPA more salary expectation espeacially when student is having 9+ GPA.

--Visualization shows us that middle class student(5-7 Lakh family income) has the highest salary expectation, than upper middle class (7+ Lakh family income) and lowest salary expectation are of financially weak students(0-5 lpa family income)

--Visualization shows us that student with 3-5 months of experince have similar salary expectation but as their experince rises to 6 months their salary expectation gets higher and remain constant till 8 months.

Que 13 : Do students in leadership positions during their college years tend to have higher GPAs or better expected salary ? 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Data analyst Data.csv')

# Check unique values in the "Leadership- skills" column
print(df['Leadership- skills'].unique())

# Calculate the average GPA and average expected salary for students with and without leadership experience
avg_gpa_leadership = df[df['Leadership- skills'] == 'yes']['CGPA'].mean()
avg_gpa_non_leadership = df[df['Leadership- skills'] == 'no']['CGPA'].mean()

avg_salary_leadership = df[df['Leadership- skills'] == 'yes']['Expected salary (Lac)'].mean()
avg_salary_non_leadership = df[df['Leadership- skills'] == 'no']['Expected salary (Lac)'].mean()

print("Average GPA for students with leadership experience:", avg_gpa_leadership)
print("Average GPA for students without leadership experience:", avg_gpa_non_leadership)
print("Average Expected Salary for students with leadership experience:", avg_salary_leadership)
print("Average Expected Salary for students without leadership experience:", avg_salary_non_leadership)

# Plotting
plt.figure(figsize=(10, 6))

# Bar plot for average GPA
plt.subplot(1, 2, 1)
plt.bar(['Leadership', 'Non-Leadership'], [avg_gpa_leadership, avg_gpa_non_leadership], color=['skyblue', 'salmon'])
plt.title('Average GPA by Leadership Status')
plt.ylabel('Average GPA')
plt.ylim(0, 10)

# Bar plot for average expected salary
plt.subplot(1, 2, 2)
plt.bar(['Leadership', 'Non-Leadership'], [avg_salary_leadership, avg_salary_non_leadership], color=['skyblue', 'salmon'])
plt.title('Average Expected Salary by Leadership Status')
plt.ylabel('Average Expected Salary (Lac)')

plt.tight_layout()
plt.show()
```

As seen in visualization results there is not much difference in average GPA and average expected salary whether Leadership quality is there or not.

Que 18: How many students know event from their college ? Which of these Top 5 colleges ?

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Data analyst Data.csv')

# Filter the dataset for students who know about the event from their college based on 'How did you come to know about this event?' column
college_counts_from_column = df[df['How did you come to know about this event?'] == 'Others']
college_counts_from_column = college_counts_from_column[college_counts_from_column['Specify in "Others" (how did you come to know about this event)'] == 'College']['College Name'].value_counts()

# Filter the dataset for students who know about the event from their college based on 'How did you come to know about this event?' column
college_counts_from_spoc = df[df['How did you come to know about this event?'] == 'SPOC/ College Professor']['College Name'].value_counts()

# Combine the counts
college_counts = college_counts_from_column.add(college_counts_from_spoc, fill_value=0)

# Print the top 5 colleges with the highest number of students who know about the event from their college
top_5_colleges = college_counts.nlargest(5)
print("Top 5 colleges with the highest number of students who know about the event from their college:")
print(top_5_colleges)
```
